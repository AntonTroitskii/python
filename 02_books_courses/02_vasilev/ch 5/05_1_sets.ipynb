```python
# Первое множество ( множество А )
A = {1, 2, 3, 4}
# Отображаем содержимое множества
print("Множество А : ", A)
# Второе множе ство ( множество В )
В = {3, 4, 5, 6}
# Отображаем содержимое множества
print("Множество В : ", В)
# Разность множеств (множество С )
С = A-В
# Отображаем результат
print("Множество С=А-В : ", С)
# Разность множеств
print("Множество A.difference(B):", A.difference(B))
print("Множество В.difference(А):", В.difference(A))
# Изменяем множество А
A.difference_update(B)
# Проверяем результат
print("Множество А : ", A)
# Изменяем множество В
В = В - {4, 6, 8, 10}
# Проверяем результат
print("Множество В : ", В)
# Изменяем множество С
С -= {1, 3, 5}
# Проверяем результат
print("Множество С : ", С)
```
